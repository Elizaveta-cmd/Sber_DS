# Альтернативный источник данных


```python
import cianparser

moscow_parser = cianparser.CianParser(location="Москва")
data = moscow_parser.get_flats(deal_type="sale", rooms=(1),\
                               with_saving_csv=True,\
                               with_extra_data = True,\
                               additional_settings={\
                                                    "start_page":1,\
                                                    "end_page":54,\
                                                    "max_price": 12000000,\
                                                 #   "metro": "Московский",\
                                                  #  "metro_station":'Раменки',\
                                                    "metro_foot_minute": 20,\
                                                    "flat_share": 2,\
                                                    "only_flat": True,\
                                                    "sort_by": "price_from_min_to_max"}
                              )
```

```python
data[0]
```

```python
import pandas as pd
import pickle
import sqlite3
```

```python
df = pd.DataFrame(data)
df.head(n=1)
```
